```python
import numpy as np
import pandas as pd
import re
import requests
```

```python
url = "https://www.flipkart.com/search?q=smartphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_5_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_2_5_na_na_ps&as-pos=2&as-type=RECENT&suggestionId=smartphone&requestId=4be2f8af-c6bb-4d3d-8a39-325b9251f15c&as-searchtext=smart"
```

```python
url
```

```python
pages=requests.get(url)
pages
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(pages.text)
soup
```

```python
soup.find_all("div",class_="RG5Slk")
```

```python
for i in soup.find_all("div",class_="RG5Slk"):
    text=i.text
    print(text)
```

```python
Brands = []

for i in soup.find_all("div", class_='RG5Slk'):
    text = i.text.strip()
    match = re.findall(r"^[A-Za-z+]+\s+[A-Za-z0-9]+",text)
    
    if match:
        clean = match[0].strip()
        Brands.append(clean)
        print(clean)
    else:
        Brands.append(np.nan)
```

```python
len(Brands)
```

```python
Network = []

for i in soup.find_all("div", class_='RG5Slk'):
    text = i.text.strip()
    
    if re.search(r"\b5G\+?\b", text):
        Network.append("5G")
    else:
        Network.append("4G")
```

```python
len(Network)
```

```python
Color = []

for i in soup.find_all("div", class_='RG5Slk'):
    text = i.text.strip()
    match = re.findall(r"\(([^,]+)", text)
    if match:
        clean = match[0].strip()
        Color.append(clean)
    else:
        Color.append(np.nan)
```

```python
len(Color)
```

```python
Ratings = []
for i in soup.find_all("div", class_='jIjQ8S'):
    div = i.find("div", class_='MKiFS6').text
    
    if div:
        Ratings.append(div)
    else:
        Ratings.append(np.nan)
```

```python
len(Ratings)
```

```python
ROM = []

for i in soup.find_all("div", class_='RG5Slk'):
    text = i.text.strip()
    match = re.findall(r"(\d+)\s*GB", text)

    if match:
        ROM.append(match[0])  
    else:
        ROM.append(np.nan)
```

```python
len(ROM)
```

```python
RAM = []

for i in soup.find_all("div", class_='col col-7-12'):
    text = i.get_text(strip=True)

    match = re.findall(r"(\d+)\s*GB\s*RAM", text, re.I)

    if match:
        RAM.append(match[0])
    else:
        RAM.append(np.nan)
```

```python
len(RAM)
```

```python
Display_type = []

for i in soup.find_all("div", class_='col col-7-12'):
    text = i.text.strip()
    match = re.findall(r"\d+(?:\.\d+)?\s*cm\s*\(\d+(?:\.\d+)?\s*inch\)\s*(?:HD\+?|Full\s*HD\+)\s*Display",text)

    if match:
        clean = match[0].strip()
        Display_type.append(clean)
    else:
        Display_type.append(np.nan)
```

```python
len(Display_type)
```

```python
CameraSpecs = []

for i in soup.find_all("div", class_='col col-7-12'):
    text = i.get_text(" ", strip=True)

    matches = re.findall(r"\d+MP(?:\s*\+\s*\d+MP)*\s*(?:Rear\s*Camera)?\s*\|\s*\d+MP",text)

    if matches:
        clean = re.sub(r"\s*\|\s*", ",", matches[0])
        CameraSpecs.append(clean)
    else:
        CameraSpecs.append(np.nan)
```

```python
len(CameraSpecs)
```

```python
Battery = []

for i in soup.find_all("div", class_='col col-7-12'):
    text = i.get_text(" ", strip=True)

    match = re.findall(r"(\d+)\s*mAh\s*Battery", text, re.I)

    if match:
        Battery.append(match[0])
    else:
        Battery.append(np.nan)
```

```python
len(Battery)
```

```python
Warranty = []

for i in soup.find_all("div", class_='col col-7-12'):
    text = i.get_text(" ", strip=True)

    match = re.findall(
        r"1\s*Year(?:\s*Manufacturer)?(?:\s*Warranty)?"
        r"(?:\s*(?:for|on)?\s*(?:Phone|Handset|Device))?"
        r"(?:\s*and\s*6\s*Months"
        r"(?:\s*(?:Manufacturer)?\s*Warranty)?"
        r"(?:\s*(?:for|on)?\s*(?:In\s*the\s*Box|Inbox)?\s*Accessories)?)?",
        text,
        re.I
    )

    if match:
        Warranty.append(match[0])
    else:
        Warranty.append(np.nan)
```

```python
len(Warranty)
```

```python
Price = []

for i in soup.find_all("div", class_='col col-5-12 mao5dl'):
    text = i.get_text(" ", strip=True)

    match = re.findall(r"₹\s*([\d,]+)", text)

    if match:
        clean_price = match[0].replace(",", "")
        Price.append(clean_price)
    else:
        Price.append(np.nan)
```

```python
len(Price)
```

```python
Brands_n = []
Network_n = []
Color_n = []
RAM_n  = []        
ROM_n  = []
CameraSpecs_n  = []
Battery_n = []     
Warranty_n = []   
Price_n = []
Ratings_n  = [] 


for i in range(1,21):
    url_n = f"https://www.flipkart.com/search?q=smartphone&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_2_5_na_na_ps&otracker1=AS_Query_OrganicAutoSuggest_2_5_na_na_ps&as-pos=2&as-type=RECENT&suggestionId=smartphone&requestId=4be2f8af-c6bb-4d3d-8a39-325b9251f15c&as-searchtext=smart&page={i}"
    page_n = requests.get(url_n,headers = {'User-Agent':'Mozilla/5.0'})
    soup_n = BeautifulSoup(page_n.text)
    
    for i in soup.find_all("div", class_='RG5Slk'):
        text = i.text.strip()
        match = re.findall(r"^[A-Za-z+]+\s+[A-Za-z0-9]+",text)
        if match:
            clean = match[0].strip()
            Brands_n.append(clean)
            print(clean)
        else:
            Brands_n.append(np.nan)
        
    for i in soup.find_all("div", class_='RG5Slk'):
        text = i.text.strip()
        if re.search(r"\b5G\+?\b", text):
            Network_n.append("5G")
        else:
            Network_n.append("4G")
    for i in soup.find_all("div", class_='RG5Slk'):
        text = i.text.strip()
        match = re.findall(r"\(([^,]+)", text)
        if match:
            clean = match[0].strip()
            Color_n.append(clean)
        else:
            Color_n.append(np.nan) 
    for i in soup.find_all("div", class_='col col-7-12'):
        text = i.get_text(strip=True)
        match = re.findall(r"(\d+)\s*GB\s*RAM", text, re.I)
        if match:
            RAM_n.append(match[0])
        else:
            RAM_n.append(np.nan)
    for i in soup.find_all("div", class_='RG5Slk'):
        text = i.text.strip()
        match = re.findall(r"(\d+)\s*GB", text)
        if match:
            ROM_n.append(match[0])  
        else:
            ROM_n.append(np.nan)
    for i in soup.find_all("div", class_='col col-7-12'):
        text = i.get_text(" ", strip=True)
        matches = re.findall(r"\d+MP(?:\s*\+\s*\d+MP)*\s*(?:Rear\s*Camera)?\s*\|\s*\d+MP",text)
        if matches:
            clean = re.sub(r"\s*\|\s*", ",", matches[0])
            CameraSpecs_n.append(clean)
        else:
            CameraSpecs_n.append(np.nan)
    for i in soup.find_all("div", class_='col col-7-12'):
        text = i.get_text(" ", strip=True)
        match = re.findall(r"(\d+)\s*mAh\s*Battery", text, re.I)
        if match:
            Battery_n.append(match[0])  
        else:
            Battery_n.append(np.nan)
    for i in soup.find_all("div", class_='col col-7-12'):
        text = i.get_text(" ", strip=True)
        match = re.findall(
            r"1\s*Year(?:\s*Manufacturer)?(?:\s*Warranty)?"
            r"(?:\s*(?:for|on)?\s*(?:Phone|Handset|Device))?"
            r"(?:\s*and\s*6\s*Months"
            r"(?:\s*(?:Manufacturer)?\s*Warranty)?"
            r"(?:\s*(?:for|on)?\s*(?:In\s*the\s*Box|Inbox)?\s*Accessories)?)?",text,re.I)
        if match:
            Warranty_n.append(match[0])
        else:
            Warranty_n.append(np.nan)
    for i in soup.find_all("div", class_='col col-5-12 mao5dl'):
        text = i.get_text(" ", strip=True)
        match = re.findall(r"₹\s*([\d,]+)", text)
        if match:
            clean_price = match[0].replace(",", "")
            Price_n.append(clean_price)
        else:
            Price_n.append(np.nan)
    for i in soup.find_all("div", class_='jIjQ8S'):
        div = i.find("div", class_='MKiFS6').text
        if div:
            Ratings_n.append(div)
        else:
            Ratings_n.append(np.nan)
```

This code creates a new column called Value_Rank that ranks devices based on their value-for-money score.
It orders all devices from best to worst based on how much performance they offer for the price.
A ranking that shows which devices offer the best value for money.

```python
df["Value_Rank"] = df["Value_Score"].rank(ascending=False)
```

```python
df["Value_Rank"] 
```

This code creates a new column called Recommendation that classifies devices based on their value-for-money performance.
It groups devices into recommendation levels based on how well they perform relative to their price compared to other devices.
A recommendation label that highlights the best value-for-money devices.

```python
def recommend(row):
    if row["Value_Score"] > df["Value_Score"].quantile(0.75):
        return "Highly Recommended"
    elif row["Value_Score"] > df["Value_Score"].quantile(0.4):
        return "Recommended"
    else:
        return "Average"

df["Recommendation"] = df.apply(recommend, axis=1)
```
